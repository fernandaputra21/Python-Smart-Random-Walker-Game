```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define The Targets
import matplotlib.patches as mpatches
targ0 = mpatches.Circle((0,0), 10, fill = None)
# The Shooter Shoots randomly
# Define maximum number of steps
n_bullets = 10
i = 0
position = []
bullets = []
while i <= n_bullets:
    pos_x = np.round(np.random.uniform(low = -10, high = 10))
    pos_y = np.round(np.random.uniform(low = -10, high = 10))
    position = [pos_x, pos_y]
    bullets.append(position)
    i += 1
bullets = np.array(bullets)
bullet_x = bullets[:,0]
bullet_y = bullets[:,1]

plt.figure(figsize=(10,10))
plt.xticks(ticks= np.arange(-10,10,1))
plt.yticks(ticks= np.arange(-10,10,1))
plt.scatter(bullet_x, bullet_y, color = 'red')
plt.gca().add_patch(targ0)
```
