```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define circle equation for restricting function
def distance(x, y):
    r_targ = np.sqrt(x**2 + y**2)
    return r_targ
# Define The Targets
import matplotlib.patches as mpatches
targ0 = mpatches.Circle((0,0), 10, fill = None)
# The Shooter Shoots randomly
# Define maximum number of steps
n_bullets = 50
i = 0
position = []
bullets = []
dist = [0]

# set initial point to shoot
x = np.random.uniform(low = -10, high = 10)
y = np.random.uniform(low = -10, high = 10)

dev = 5 # Set initial deviation

while i <= n_bullets:
    pos_x = np.random.normal(loc = x, scale = 1)
    pos_y = np.random.normal(loc = y, scale = 1)
    position = [pos_x, pos_y]
    bullets.append(position)
    dist.append(distance(pos_x, pos_y))
    if dist[i+1] < dist[i]:
        x = pos_x
        y = pos_y
        dev = np.random.normal(loc = 0, scale = 1.0) # restricting the deviation from the previous shot
    if dist[i+1] > dist[i]:
        x = x
        y = y
        dev = np.random.normal(loc = 0, scale = 2.0) # increase the deviation from the previous shot
   # print([pos_x, pos_y])
    i += 1
bullets = np.array(bullets)
bullet_x = bullets[:,0]
bullet_y = bullets[:,1]
plt.figure(figsize=(10,10))
plt.xticks(ticks= np.arange(-10,10,1))
plt.yticks(ticks= np.arange(-10,10,1))
plt.scatter(bullet_x, bullet_y, color = 'red')
plt.gca().add_patch(targ0)
print('Distance from center', np.round(np.array(dist)[1:]))
print('The distance from the center is supposed to be smaller for every shot')
```
