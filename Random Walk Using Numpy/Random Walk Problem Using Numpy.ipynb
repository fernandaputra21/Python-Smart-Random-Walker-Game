```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# The walker will walk always go up along the y axis but can move in the x axis randomly
x = 0
y = 0
arr_x = []
arr_y = []
# Define maximum number of steps
n_steps = 100
i = 0
while i <= n_steps:
    step_x = np.round(np.random.uniform(low = -1.0, high = 1.0))
    x+= step_x
    y+= 1
    arr_x.append(x)
    arr_y.append(y)
    i += 1
#print(arr_x, arr_y)
arr_x = np.array(arr_x)
arr_y = np.array(arr_y)
plt.plot(arr_x, arr_y)
```
