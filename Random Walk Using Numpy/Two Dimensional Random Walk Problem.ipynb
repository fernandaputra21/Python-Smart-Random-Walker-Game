```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# The walker will walk always go up along the y axis but can move in the x axis randomly
x = 0
y = 0
arr_x = []
arr_y = []
# Define maximum number of steps
n_steps = 100
i = 0
while i <= n_steps:
    step_x = np.round(np.random.uniform(low = -1.0, high = 1.0))
    step_y = np.round(np.random.uniform(low = -1.0, high = 1.0))
    x+= step_x
    y+= step_y
    arr_x.append(x)
    arr_y.append(y)
    i += 1
#print(arr_x, arr_y)
arr_x = np.array(arr_x)
arr_y = np.array(arr_y)
plt.plot(arr_x, arr_y)
```

```python
# Random Walk inside a wall

# Define a wall
import matplotlib.patches as mpatches

wall = mpatches.Rectangle((-5,-5), 10, 10, fill = False)

x = 0
y = 0
arr_x = []
arr_y = []
# Define maximum number of steps
n_steps = 1000
i = 0
while i <= n_steps:
    step_x = np.round(np.random.uniform(low = -1.0, high = 1.0))
    step_y = np.round(np.random.uniform(low = -1.0, high = 1.0))
    x+= step_x
    y+= step_y
    arr_x.append(x)
    arr_y.append(y)
    i += 1
    if x == 5:
        step_x = -1
        x+= step_x
        arr_x.append(x)
        arr_y.append(y)
        i += 1
    elif x == -5:
        step_x = 1
        x+= step_x
        arr_x.append(x)
        arr_y.append(y)
        i += 1
    if y == 5:
        step_y = -1
        y+= step_y
        arr_x.append(x)
        arr_y.append(y)
        i += 1
    if y == -5:
        step_y = 1
        y+= step_y
        arr_x.append(x)
        arr_y.append(y)
        i += 1
#print(arr_x, arr_y)
arr_x = np.array(arr_x)
arr_y = np.array(arr_y)
plt.figure(figsize=(10,10))
plt.xticks(ticks= np.arange(-10,10,1))
plt.yticks(ticks= np.arange(-10,10,1))
plt.plot(arr_x, arr_y)
plt.title('Random Walker in a Box')
plt.gca().add_patch(wall)
```
